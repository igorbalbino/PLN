```python
def remove_brackets(column):
    res = []
    for w in column:
        res.append(re.sub(r'[\\]', '', w))
    return res
print('created remove_brackets!')
```

```python
from nltk.corpus import stopwords

stopwords_pt = stopwords.words('portuguese')
stopwords_en = stopwords.words('english')

def filter_sentences(line):
    res = []
    for w in line:
        if w is not None:
            if w not in stopwords_pt and w not in stopwords_en:
                res.append(w)
    return res
                
print('created filter_sentences!')
```

```python
from nltk.tokenize import word_tokenize

def tokenize_word(sentence):
    res = []
    for w in sentence:
        if w is not None and type(w) == str:
            res.append(word_tokenize(w))
    return res
print('created tokenize_word!')
```
